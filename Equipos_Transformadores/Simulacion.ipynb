```python
import numpy as np
import matplotlib.pyplot as plt

# rango de datos de 0 a 6 con saltos de 0.1
x = np.arange(0, 6, 0.1)
# funcion seno
y1 = np.sin(x)
# funcion coseno
y2 = np.cos(x)
# graficar
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos") # lineas punteadas
plt.xlabel("x") # etiqueta eje x
plt.ylabel("y") # etiqueta eje y
plt.title('sin & cos') # titulo
plt.show()
```
