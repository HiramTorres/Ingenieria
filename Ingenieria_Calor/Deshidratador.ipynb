### Radiación solar global
De entre las 6am y las 6pm tenemos 300 y 9800 $ \frac{Wh}{m^2}$ 
Considerando una emisividad de 0.94 

$Q = (\epsilon)(A)(T)^4$
$Q= (0.4)(5.6x10-8)(298.15)^4$
$Q = 63.05$

**Energía Solar** = $550 \frac{W}{m^2}$

1 $m^2$ = $550 W$ \
0.16 $m^2$ = $88 W$

```python
from turtle import color
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10,6))
y = [37,37.5,38,38.3,39.6,39]
y_1 = [31,32.6,30.8,30.8,30,29]
x = ["0","20","40","60","80","100"]
plt.bar(x, y, color = "c")
plt.bar(x, y_1, color = "m")
plt.plot(x, y, color = "r", marker = "o", linestyle = "dashdot")
plt.plot(x, y_1, color = "k", marker = "o", linestyle = "dashdot")
plt.xlabel("Tiempo (min)")
plt.ylabel("Temperatura (°C)")
plt.title("Transferencia de calor en parrilla")
plt.show()
```

```python
fig, ax = plt.subplots(2,1, sharey = True)
ax[0].plot(x,y,color = 'm',label = 'T Parrilla')
ax[0].legend()
ax[1].plot(x,y_1,color = 'b',label = 'T inicial')
ax[1].legend()
plt.xlabel("Tiempo (min)")
plt.ylabel("Temperatura (°C)")

plt.show()
```
