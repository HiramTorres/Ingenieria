```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(0, 10)
y = 1.8 / x - 0.2


plt.figure(figsize= (6,6))
plt.plot(x,y, linestyle = "--", color = "c")
plt.xlabel("Volumen")
plt.ylabel("Presión")
plt.xlim(0,3)
plt.ylim(0,3)
plt.yticks([1,2])
plt.xticks([1,2])
plt.legend(loc= "upper right")
plt.annotate("Proceso adiábtico", xy=(1.5,1.5),textcoords='offset points')
plt.savefig("Proceso_adiabatico.jpg")

plt.show()
```
